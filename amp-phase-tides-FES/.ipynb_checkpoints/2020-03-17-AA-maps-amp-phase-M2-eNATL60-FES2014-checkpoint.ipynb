```python
import sys
import numpy as np
import xarray as xr
import numpy.ma as ma
from netCDF4 import Dataset

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as mplcm

import cartopy.crs as ccrs
import matplotlib.ticker as mticker

from decimal import Decimal

%matplotlib inline
```

```python
def one_plot_cont_natl60(ax,freq,var,var2,varname,vmin,vmax,cmap):
    ds=xr.open_dataset('/scratch//eNATL60-BLBT01_y2009m02d01-m03d16_harm_ssh_tidal_0-360.nc')
    lat=ds['nav_lat']
    lon=ds['nav_lon']
    amp=ds[var]
    ph=ds[var2]
    
    ax.coastlines(resolution="10m")

    ax.set_extent([-100,50,0,60])

    gl = ax.gridlines(crs=pj, draw_labels=True, linewidth=1, color='black', alpha=0.2, linestyle='--')
    label_style= {'size' : 12, 'color' : 'black'}
    gl.xlabel_style = label_style
    gl.xlabels_top = False
    gl.xlocator = mticker.FixedLocator(np.arange(-110,100,20,dtype=float))
    gl.ylabel_style = label_style
    gl.ylabels_right = False
    gl.ylocator = mticker.FixedLocator(np.arange(-20,90,10,dtype=float))
    cs = plt.pcolormesh(lon, lat, amp,transform=ccrs.PlateCarree(),cmap=cmap,vmin=vmin,vmax=vmax)
#    plt.contour(lon,lat,ph,[-180,-150,-120,-90,-60,-30,0,30,60,90,120,150,180],cmap=cmocean.cm.phase,linestyle='solid',linewidths=1,transform=ccrs.PlateCarree())
    plt.contour(lon,lat,ph,[30,60,90,120,150,180,210,240,270,300,330],cmap=cmocean.cm.phase,linestyle='solid',linewidths=1,transform=ccrs.PlateCarree())
    plt.contour(lon,lat,ph,[1],colors='k',linestyle='solid',linewidths=1,transform=ccrs.PlateCarree())

    plt.title(freq+' '+varname+' eNATL60')
    cbar = plt.colorbar(cs,orientation='horizontal',shrink=0.75,label='m')
    
```

```python
def one_plot_cont_fes(ax,freq,var,var2,varname,vmin,vmax,cmap):
    ds=xr.open_dataset('/Users/auraoupa/Data/FES2014/'+freq+'.FES2014b_elevation_RI_AP_0-360.nc')
    lon=ds['lon']
    lat=ds['lat']
    amp=ds[var]
    ph=ds[var2]
    mask=amp>0
    mask_one=mask.astype(int)
    amp_ma=np.ma.masked_where(1-mask_one,amp)
    mask_phi=ph!=45
    mask_one_phi=mask_phi.astype(int)
    ph_ma=np.ma.masked_where(1-mask_one_phi,ph)

    ax.coastlines(resolution="10m")

    ax.set_extent([-100,50,0,60])

    gl = ax.gridlines(crs=pj, draw_labels=True, linewidth=1, color='black', alpha=0.2, linestyle='--')
    label_style= {'size' : 12, 'color' : 'black'}
    gl.xlabel_style = label_style
    gl.xlabels_top = False
    gl.xlocator = mticker.FixedLocator(np.arange(-110,100,20,dtype=float))
    gl.ylabel_style = label_style
    gl.ylabels_right = False
    gl.ylocator = mticker.FixedLocator(np.arange(-20,90,10,dtype=float))
    cs = plt.pcolormesh(lon, lat, amp_ma,transform=ccrs.PlateCarree(),cmap=cmap,vmin=vmin,vmax=vmax)
#    plt.contour(lon,lat,ph_ma,[-180,-150,-120,-90,-60,-30,0,30,60,90,120,150,180],cmap=cmocean.cm.phase,linestyle='solid',linewidths=1,transform=ccrs.PlateCarree())
    plt.contour(lon,lat,ph_ma,[30,60,90,120,150,180,210,240,270,300,330],cmap=cmocean.cm.phase,linestyle='solid',linewidths=1,transform=ccrs.PlateCarree())
    plt.contour(lon,lat,ph_ma,[1],colors='k',linestyle='solid',linewidths=1,transform=ccrs.PlateCarree())
    plt.title(freq+' '+varname+' FES2014')

    cbar = plt.colorbar(cs,orientation='horizontal',shrink=0.75,label='m')
```

```python
cmap=plt.get_cmap('rainbow')
matplotlib.rcParams['contour.negative_linestyle']= 'solid'
matplotlib.rcParams['lines.linewidth']= 2

fig=plt.figure(figsize=(21,40))
pj= ccrs.PlateCarree()

ax=plt.subplot(5,2,1,projection=pj)
one_plot_cont_natl60(ax,'M2','M2_A','M2_G','amplitude',0,2,cmap)

ax=plt.subplot(5,2,2,projection=pj)
one_plot_cont_fes(ax,'M2','r','phi','amplitude',0,2,cmap)

ax=plt.subplot(5,2,3,projection=pj)
one_plot_cont_natl60(ax,'S2','S2_A','S2_G','amplitude',0,1,cmap)

ax=plt.subplot(5,2,4,projection=pj)
one_plot_cont_fes(ax,'S2','r','phi','amplitude',0,1,cmap)

ax=plt.subplot(5,2,5,projection=pj)
one_plot_cont_natl60(ax,'N2','N2_A','N2_G','amplitude',0,0.5,cmap)

ax=plt.subplot(5,2,6,projection=pj)
one_plot_cont_fes(ax,'N2','r','phi','amplitude',0,0.5,cmap)

ax=plt.subplot(5,2,7,projection=pj)
one_plot_cont_natl60(ax,'K1','K1_A','K1_G','amplitude',0,0.5,cmap)

ax=plt.subplot(5,2,8,projection=pj)
one_plot_cont_fes(ax,'K1','r','phi','amplitude',0,0.5,cmap)

ax=plt.subplot(5,2,9,projection=pj)
one_plot_cont_natl60(ax,'O1','O1_A','O1_G','amplitude',0,0.5,cmap)

ax=plt.subplot(5,2,10,projection=pj)
one_plot_cont_fes(ax,'O1','r','phi','amplitude',0,0.5,cmap)

plt.savefig('/Users/auraoupa/Data/PLOTS/maps_amp_phi_M2N2S2K1O1_eNATL60-FES2014.png')
```
